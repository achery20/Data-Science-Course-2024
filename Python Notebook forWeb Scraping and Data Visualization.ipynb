Step 2: Clean the data by formatting the dates of the reviews and removing unnecessary fields. Convert the filtered data into JSON strings. Each review data will contain the title, body, rating, and date.

```python
import json
from datetime import datetime

# Load JSON data from a file
with open("/content/amazonreviews_scraping_visualizations.json", "r") as file:
    json_data = json.load(file)

# Access the "reviews" key to iterate over reviews
reviews = json_data["reviews"]

# Filter the reviews to include only the required fields and format the date
filtered_reviews = []
for review in reviews:
    # Parse the date
    date_raw = review["date"]["raw"]
    parsed_date = datetime.strptime(date_raw, "Reviewed in the United States on %B %d, %Y")

    # Format the date as YYYY-MM-DD
    formatted_date = parsed_date.strftime("%Y-%m-%d")

    filtered_review = {
        "title": review.get("title"),
        "body": review.get("body"),
        "rating": review.get("rating"),
        "date": formatted_date
    }
    filtered_reviews.append(filtered_review)

# Convert the filtered review data to JSON string
filtered_reviews_json = json.dumps(filtered_reviews, indent=4)

# Print the JSON string
print(filtered_reviews_json)
```

Step 3: Create the data visualizations

```python
Bar plot that shows the rating distributions of the reviews
```

```python
import json
import matplotlib.pyplot as plt

# Load JSON data from a file
with open("/content/amazonreviews_scraping_visualizations.json", "r") as file:
    json_data = json.load(file)

# Access the "reviews" key to iterate over reviews
reviews = json_data["reviews"]

# Extract ratings from the reviews
ratings = [review["rating"] for review in reviews]

# Calculate the frequency of each rating
rating_counts = {}
for rating in ratings:
    if rating in rating_counts:
        rating_counts[rating] += 1
    else:
        rating_counts[rating] = 1

# Sort the ratings by value
sorted_ratings = sorted(rating_counts.items())

# Extract the ratings and corresponding frequencies
ratings, frequencies = zip(*sorted_ratings)

# Create the bar plot for rating distribution
plt.bar(ratings, frequencies, color='skyblue')

# Add labels and title
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.title('Rating Distribution of Reviews')

plt.savefig('rating_distribution.png')
# Show plot
plt.show()
```

Line graph that shows the frequency of reviews over time

```python
import json
import matplotlib.pyplot as plt
from collections import defaultdict

# Load JSON data from a file
with open("/content/amazonreviews_cleaned_visualizations.json", "r") as file:
    json_data = json.load(file)

# Extract dates and ratings from the reviews
reviews = json_data
dates = [review["date"] for review in reviews]

# Count the number of reviews for each date
review_counts = defaultdict(int)
for date in dates:
    review_counts[date] += 1

# Sort the dates
sorted_dates = sorted(review_counts.keys())

# Extract the sorted dates and corresponding review counts
sorted_review_counts = [review_counts[date] for date in sorted_dates]

# Plot the number of reviews over time
plt.figure(figsize=(10, 6))
plt.plot(sorted_dates, sorted_review_counts, marker='o', linestyle='-')

# Add labels and title
plt.xlabel('Date')
plt.ylabel('Number of Reviews')
plt.title('Number of Reviews Over Time')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

plt.savefig('reviews_over_time.png')

# Show plot
plt.tight_layout()
plt.show()
```

A word cloud that shows the main focus of the reviews. Stop words were removed and lemmatization was used on the data to ensure the word cloud contained meaningful information.

```python
import json
from wordcloud import WordCloud
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import string
import re
import matplotlib.pyplot as plt
import nltk

# Download NLTK stopwords and WordNet data
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Load JSON data from a file
with open("/content/amazonreviews_cleaned_visualizations.json", "r") as file:
    json_data = json.load(file)

# Extract review bodies from the JSON data
reviews = json_data
review_bodies = [review["body"] for review in reviews]

# Combine all review bodies into a single string
all_text = ' '.join(review_bodies)

# Define additional words to remove
additional_stopwords = set(['phone', 'smartphone', 'device', 'samsung', 'galaxy', 'ultra', 's','really','s24','lot','something','thing','user','want','without','use','one','said'])

# Tokenize the text
tokens = word_tokenize(all_text)

# Remove stopwords, additional words, single-letter tokens, tokens with apostrophes, and punctuation
stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word.lower() not in stop_words
                   and word.lower() not in additional_stopwords
                   and len(word) > 1
                   and "'" not in word
                   and word not in string.punctuation]

# Lemmatization
lemmatizer = WordNetLemmatizer()
lemmatized_tokens = [lemmatizer.lemmatize(word) for word in filtered_tokens]

# Combine tokens back into a string
processed_text = ' '.join(lemmatized_tokens)

# Generate a word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(processed_text)

# Plot the word cloud
plt.figure(figsize=(10, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.title('Word Cloud of Reviews')
plt.axis('off')
plt.savefig('word_cloud_reviews.png')
plt.show()
```

A bar graph shows the sentiment distribution of certain features of the Galaxy S24 such as the camera quality andbattery life. Sentiment analysis was performed to analyze the data.

```python
import json
import matplotlib.pyplot as plt
from textblob import TextBlob

# Load JSON data from a file
with open("/content/amazonreviews_cleaned_visualizations.json", "r") as file:
    json_data = json.load(file)

# Extract review bodies from the JSON data
reviews = json_data
review_bodies = [review["body"] for review in reviews]

# Define the features you want to analyze
features = ["screen", "camera", "battery", "speaker", "price", "display", "security"]

# Initialize sentiment dictionaries for each feature
feature_sentiments = {feature: {"positive": 0, "negative": 0, "neutral": 0} for feature in features}

# Perform sentiment analysis on the review bodies
for body in review_bodies:
    for feature in features:
        if feature in body.lower():
            analysis = TextBlob(body)
            polarity = analysis.sentiment.polarity
            if polarity > 0.2:
                feature_sentiments[feature]["positive"] += 1
            elif polarity < -0.2:
                feature_sentiments[feature]["negative"] += 1
            else:
                feature_sentiments[feature]["neutral"] += 1

# Plot the sentiment distribution for each feature
plt.figure(figsize=(10, 6))
for feature, sentiments in feature_sentiments.items():
    labels = list(sentiments.keys())
    counts = list(sentiments.values())
    plt.bar(labels, counts, alpha=0.7, label=feature)

plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.title('Sentiment Distribution for Features')
plt.legend()
plt.savefig('sentiment_distribution.png')
plt.show()
```

A bar plot that shows the sentiment polarity distribution of the reviews. The reviews leaned towards a positive polaritty which shows that most customers were satisfied with the product.

```python
import json
import matplotlib.pyplot as plt
from textblob import TextBlob

# Load JSON data from a file
with open("/content/amazonreviews_cleaned_visualizations.json", "r") as file:
    json_data = json.load(file)

# Extract review bodies from the JSON data
reviews = json_data
review_bodies = [review["body"] for review in reviews]

# Perform sentiment analysis on the review bodies
polarities = [TextBlob(body).sentiment.polarity for body in review_bodies]

# Plot the sentiment polarity distribution
plt.figure(figsize=(10, 6))
plt.hist(polarities, bins=20, color='skyblue', edgecolor='black', alpha=0.7)
plt.xlabel('Polarity')
plt.ylabel('Frequency')
plt.title('Sentiment Polarity Distribution')
plt.savefig('sentiment_polarity_distribution.png')
plt.show()
```
